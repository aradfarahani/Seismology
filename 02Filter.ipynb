```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from obspy import read
```

```python
st = read("RJOB_061005_072159.ehz.new")
```

```python
print(st)
```

```python
tr = st[0]
```

```python
plt.figure(figsize=(10, 4))
plt.plot(tr.times(), tr.data, label='Raw Data')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Seismic Data')
plt.legend()
plt.show()
```

```python
tr_filtered = tr.copy()
tr_filtered.filter('bandpass', freqmin=1.0, freqmax=10.0)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(tr_filtered.times(), tr_filtered.data, label='Filtered Data', color='orange')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Filtered Seismic Data')
plt.legend()
plt.show()
```
