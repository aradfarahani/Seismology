```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Generate synthetic geophysics data

```python
def generate_synthetic_data(num_points=100):
    depth = np.linspace(0, 1000, num_points)
    velocity = 1500 + 100 * np.sin(depth / 100)  # Synthetic velocity data
    data = pd.DataFrame({'Depth': depth, 'Velocity': velocity})
    return data
```

## Compute basic statistics

```python
def compute_statistics(data):
    stats = {
        'mean_velocity': np.mean(data['Velocity']),
        'max_velocity': np.max(data['Velocity']),
        'min_velocity': np.min(data['Velocity']),
        'std_velocity': np.std(data['Velocity'])
    }
    return stats
```

## Plot geophysics data

```python
def plot_geophysics_data(data):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Depth'], data['Velocity'], label='Velocity', color='blue')
    plt.xlabel('Depth (m)')
    plt.ylabel('Velocity (m/s)')
    plt.title('Velocity vs. Depth')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Generate portfolio

```python
def generate_geophysics_portfolio():
    data = generate_synthetic_data()
    stats = compute_statistics(data)
    print("Basic Statistics:")
    for key, value in stats.items():
        print(f"{key}: {value:.2f}")
    plot_geophysics_data(data)
```

```python
generate_geophysics_portfolio()
```
